```python
import pandas as pd
import numpy as np
from googletrans import Translator #serve para traduzir os nomes dos paises
import ast
import re
```

# Importa Dados do Transparência Internacional e do World Happiness Report

```python
#atenção, o arquivo excel possui mais de uma planilha
df_felicidade = pd.read_excel('http://worldhappiness.report/wp-content/uploads/sites/2/2017/03/online-data-chapter-2-whr-2017.xlsx',sheetname=2)
```

```python
df_corrupção = pd.read_excel('http://files.transparency.org/content/download/2155/13635/file/CPI2016_FullDataSetWithRegionalTables.xlsx')
```

```python
print(df_felicidade.columns)
df_felicidade.head(20)
```

```python
print(df_corrupção.columns)
df_corrupção.head()
```

```python
df_felicidade = df_felicidade[['Country','Happiness score']]
```

```python
df_corrupção = df_corrupção[['Country','Rank']]
```

# Verificar tamanho dos indexes, conserta os nomes que estão diferentes e os junta.

```python
"Belize","North Cyprus", "Palestinian Territories" #não aparecem na outra lista 
    
dic_correção_países = {
    "United States" : "The United States of America",
    "Taiwan Province of China" : "Taiwan",
    "South Korea" : "Korea (South)",
    "Hong Kong S.A.R., China" : "Hong Kong",
    "Macedonia" : "The FYR of Macedonia",
    'Congo (Brazzaville)' : "Republic of Congo" , 
    'Congo (Kinshasa)' : "The Democratic Republic of Congo",
    "Ivory Coast":"Côte d’Ivoire"
}
```

```python
for key in dic_correção_países:
    print(key,dic_correção_países[key])
    df_felicidade['Country'][df_felicidade['Country'].isin([key])] = dic_correção_países[key]
```

```python
# confere os que não estavam nas duas listas.
df_felicidade[df_felicidade['Country'].isin(df_corrupção['Country']) == False]
```

```python
#transforma o score em ranking.
df_felicidade['Happiness score'] = range(1,len(df_felicidade)+1)
df_felicidade.head()
```

```python
#junta os dataframes.
df = pd.merge(df_corrupção,df_felicidade, on='Country')
df.columns = ['país','percepção de corrupção','nível de felicidade']
len(df)
```

```python
df
```

# Faz a tradução do nomes dos países com base no pacote googltrans 

```python
translator = Translator()
```

```python
traduções = translator.translate(df['país'].tolist(),src='en',dest='pt')
```

```python
traduções = [x.text for x in traduções]
df['extra'] = traduções
```

```python
with pd.option_context("display.max_rows", 1000):
    display(df[['país','extra']])
```

```python
# alguns nomes ficaram errados.
traduções = ["Togo" if x=="Ir" else x for x in traduções]
traduções = str(traduções)
traduções = re.sub("Peru","Turquia",traduções,count=1) #traduziu Turkey=Peru rs
traduções= ast.literal_eval(traduções)

df['país'] = traduções
df = df.drop(['extra'],axis=1)
```

```python
#confere os nomes
with pd.option_context("display.max_rows", 1000):
    display(df)
```

```python
#salva em formato csv.
df.to_csv('paises.csv', index=False,encoding='utf8')
```
